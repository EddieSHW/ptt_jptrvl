```python
from ckip_transformers.nlp import CkipWordSegmenter, CkipPosTagger, CkipNerChunker
from collections import Counter
import networkx as nx
import matplotlib.pyplot as plt
```

```python
text = [
        '經過多年激烈戰事，複製人大戰即將結束。絶地議會派歐比王將導致戰亂的主謀者繩之以法；不料，西斯勢力已悄悄深入銀河系，勢力漸大的議長白卜庭用黑暗勢力的力量，誘惑天行者安納金轉變成黑武士達斯維達，幫助他達成心願建立銀河帝國，剷除絕地武士…【星際大戰】系列電影最後一塊拼圖，喬治盧卡斯不僅要解開黑武士的影壇跨世紀謎團，更要著手打造影史最大星際戰爭。',
        '故事由《魔戒首部曲：魔戒現身》展開，一位名叫佛羅多的年輕人，無意間得到一枚有著神秘力量的戒指，卻發現這枚戒指原來是黑暗魔王索倫所擁有。幾經波折後，他決定摧毀魔戒，以免索倫奪回去鞏固自己的勢力。巫師甘道夫、精靈、矮人、哈比人和人類於是組成魔戒遠征隊，協助佛羅多前往索倫統治的「中土世界」，將魔戒丟入末日火山摧毀。旅途中索倫派出怪獸追殺佛羅多一群人，魔戒也開始腐蝕人心，讓人產生難以抵擋的慾望，考驗著每位接觸戒指的意志力。',
        '故事講述二次大戰期間，主角史提芬羅傑斯想參軍替美國打倒納粹德軍，可是他未能通過體能檢查，於是他參加了軍方一個秘密實驗計劃—重生計劃，參加者會被改造成為超級戰士，主角史提芬被改造後成為了唯一成功的實驗品，自此之後，他身穿紅、白、藍三色戰鬥服為國效力，就成為了美國隊長。'
    ]
```

```python

# Initialize drivers
print("Initializing drivers ... WS")
ws = CkipWordSegmenter(model="bert-base", device=0)
print("Initializing drivers ... POS")
pos = CkipPosTagger(model="bert-base", device=0)
print("Initializing drivers ... NER")
ner = CkipNerChunker(model="bert-base", device=0)
print("Initializing drivers ... all done")
print()

# 中文字體
font_path = '/usr/share/fonts/google-noto-cjk/NotoSansCJK-Regular.ttc'
```

```python
def should_add_edge(noun1, noun2, co_occurrence_matrix, threshold):
    # 假設 co_occurrence_matrix 是一個詞彙共現矩陣，包含了名詞之間的共現頻率
    # threshold 是一個閾值，表示共現頻率高於該閾值時才添加邊

    # 檢查名詞1和名詞2之間的共現頻率是否高於閾值
    if co_occurrence_matrix[noun1][noun2] > threshold:
        return True
    else:
        return False
```

```python
# 使用 CKIP 分詞器進行斷詞
word_segmented_text = ws(text)

# 使用 CKIP 詞性標註器進行詞性標註
pos_tags = pos(word_segmented_text)

# 提取名詞
nouns = [word for word, pos_list in zip(word_segmented_text[0], pos_tags[0]) if
             any(pos.startswith("N") for pos in pos_list)]

# 計算詞頻
word_freq = Counter(nouns)

# 獲取詞頻最高的20個詞
top_words = word_freq.most_common(20)

# 將名詞列表轉換為空格分隔的字串
nouns_text = ' '.join(nouns)
```

```python
# 創建一個空的共現頻率字典
co_occurrence_matrix = {}

# 初始化共現頻率字典
for noun1 in nouns:
    co_occurrence_matrix[noun1] = {}
    for noun2 in nouns:
        co_occurrence_matrix[noun1][noun2] = 0

# 更新共現頻率字典
for i in range(len(nouns)):
    for j in range(i + 1, len(nouns)):
        noun1 = nouns[i]
        noun2 = nouns[j]
        # 在這裡根據文本數據更新共現頻率字典，例如，如果名詞1和名詞2在同一文本中出現，則增加共現頻率
        co_occurrence_matrix[noun1][noun2] += 1
        co_occurrence_matrix[noun2][noun1] += 1
```

```python
 # 創建一個空的無向圖
G = nx.Graph()

# 在共位詞網路中添加名詞作為節點
G.add_nodes_from(nouns)

# 使用名詞之間的共同出現頻率來添加邊
for i, noun1 in enumerate(nouns):
    for j, noun2 in enumerate(nouns):
        if i < j:
            # 在這裡添加您的邊的邏輯，根據共同出現的頻率決定是否添加邊
            if should_add_edge(noun1, noun2, co_occurrence_matrix, 25):
                G.add_edge(noun1, noun2)

# 繪製共位詞網路圖
pos_net = nx.spring_layout(G, k=0.04)  # 使用 Spring layout 布局
plt.figure(figsize=(12, 12))
nx.draw(G, pos_net, node_color='lightblue', with_labels=True, node_size=300, font_family='Noto Sans CJK JP',
            font_size=10)

plt.title('Co-occurrence Word Network')
plt.axis('off')
plt.show()
```

```python
# 共起ネットワークを作成する
G = nx.Graph()
nodes = sorted(set([j for pair in jaccard_dict.keys() for j in pair]))
G.add_nodes_from(nodes)
print('Number of nodes=', G.number_of_nodes())
for pair, coef in jaccard_dict.items():
    G.add_edge(pair[0], pair[1], weight=coef)
print('Number of edges=', G.number_of_edges())
plt.figure(figsize=(15, 15))
seed = 0
np.random.seed(seed)
pos = nx_agraph.graphviz_layout(G, prog='neato', args='-Goverlap="scalexy" -Gsep="+6" -Gnodesep=0.8 -Gsplines="polyline" -GpackMode="graph" -Gstart={}'.format(seed))
pr = nx.pagerank(G)
nx.draw_networkx_nodes(G, pos, node_color=list(pr.values()), cmap=plt.cm.rainbow, alpha=0.7, node_size=[100000*v for v in pr.values()])
nx.draw_networkx_labels(G, pos, font_family='MS Gothic', font_weight='bold')
edge_width = [d['weight'] * 8 for (u, v, d) in G.edges(data=True)]
nx.draw_networkx_edges(G, pos, alpha=0.7, edge_color='darkgrey', width=edge_width)
plt.axis('off')
plt.tight_layout()
plt.savefig('co-occurance.png', bbox_inches='tight')
```
